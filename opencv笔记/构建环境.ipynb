### 图像基本操作
* 图像有灰度图，一般只有一个通道，0-255从黑到白，也可以理解成从暗到亮
* 但是如果是RGB图那就有三个通道，分别是红绿蓝，也是分别范围是0-255
* RGB 矩阵的大小和图像的大小一样，若图像的像素点宽 500 长 500，则得到 [500,500,3] 的矩阵，这里的 3 就是 3 通道。
